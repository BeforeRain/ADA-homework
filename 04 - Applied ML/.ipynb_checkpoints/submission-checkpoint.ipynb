### Data for task1 data_agg
First, extract columns that need to be summed over(the statistics about games and cards). Using playerShort as the unique ID of each player.

```python
# add racism features somehow --> average?
# extract columns that need to be summed over
soccer_data_trim = soccer_data_clean[["playerShort", "games", "victories","ties","defeats","goals","yellowCards","yellowReds","redCards"]]
soccer_data_trim.shape
#soccer_data_trim.head()
```

```python
data_agg = soccer_data_trim.groupby("playerShort").aggregate(np.sum)
data_agg.head()
#data_agg.shape
```

Then add new columns of weighted sum of IAT and Exp scores. The weight depends on the porption of total games.

```python
ref_bias = soccer_data_clean[["playerShort", "games","meanIAT","meanExp"]].groupby("playerShort")
#ref_bias.head()
```

```python
# weighted sum of IAT and Exp scores
def weighted_average(group, feature):
    weights = group['games']
    total_games = weights.sum()
    iat_scores = group[feature]
    return (iat_scores * weights).sum() / total_games

data_agg["weightedIAT"] = ref_bias.apply(weighted_average, 'meanIAT')   
data_agg["weightedExp"] = ref_bias.apply(weighted_average, 'meanExp') 
data_agg.head()
```

Number of unique referees which gave each card type.

```python
unique_ref = soccer_data_trim.groupby("playerShort")
```

```python
# count number of non-zero entries, i.e. number of unique referees who have given the card
def num_unique_ref_card(group, card_type):
    ref_card = group[card_type]
    return (ref_card!=0).sum()

# count number of unique referees that have given a card to a particular player
def num_unique_ref(group):
    ref_cards = group['yellowCards']+group['yellowReds']+group['redCards']
    return (ref_cards!=0).sum()


data_agg['uniqueYellow'] = unique_ref.apply(num_unique_ref_card, 'yellowCards')
data_agg['uniqueYellowReds'] = unique_ref.apply(num_unique_ref_card, 'yellowReds')
data_agg['uniqueReds'] = unique_ref.apply(num_unique_ref_card, 'redCards')
data_agg['uniqueRefCards'] = unique_ref.apply(num_unique_ref)
data_agg.head()
```

```python
data_agg.tail()
```

Now we need to extract the characteristic data of the players: Name, Height, Weight, Birthday, Position and Skin Color ratings, Club, Country of the League.

```python
def extract_param(group, param):
    return group[param].values[0]

const_param = ["player","height","weight","club","leagueCountry","birthday","position","rater1","rater2"]
for param in const_param:
    data_agg[param] = soccer_data_clean.groupby("playerShort").apply(extract_param, param)

data_agg.head()
```

```python
data_agg.columns.values
```

```python
data_agg = data_agg[["player", "height", "weight", "rater1", "rater2", "club", "leagueCountry", 
                     "birthday", "position", "games", "victories", "ties", "defeats", "goals", 'yellowCards',
                     'yellowReds', 'redCards', 'weightedIAT', 'weightedExp', 'uniqueYellow', 'uniqueYellowReds', 
                     'uniqueReds', 'uniqueRefCards']]
data_agg.ix[:5,:14]
```

Create dummy variables, cross-validation

To evaluate the skin color, we use another new feature 'color_rate' which is the average value of rater1 and rater2.

```python
data_skin = soccer_data_clean[["playerShort", "rater1", "rater2"]].groupby("playerShort")
```

```python
#  average value of rater1 and rater2
def rate_average(group):
    rate = (group['rater1'].values[0]+ group['rater2'].values[0])/2
    return rate

data_agg["color_rate"] = data_skin.apply(rate_average)   
data_agg.head()
#data_agg.shape
```

Suppose there are two different skin colors: 0-white, 1-black.

```python
def skin_label(group):
    rate = (group['rater1'].values[0]+ group['rater2'].values[0])/2
    if rate < 0.5 :
        skin =0
    else:
        skin =1
    return skin

data_agg["skin_color"] = data_skin.apply(skin_label)
data_agg = data_agg.drop('player', 1)
data_agg.head()
```

Features like club, leagurCountry, birthday and position are string or object types, which should be converted to value (int/float) at first.

```python
## reference: https://gist.github.com/ghl3/11102044
def get_nominal_integer_dict(nominal_vals):
    d = {}
    for val in nominal_vals:
        if val not in d:
            current_max = max(d.values()) if len(d) > 0 else -1
            d[val] = current_max+1
    return d


def convert_to_integer(srs):
    d = get_nominal_integer_dict(srs)
    return srs.map(lambda x: d[x])


def convert_strings_to_integer(df):
    ret = pd.DataFrame()
    for column_name in df:
        column = df[column_name]
        #if column.dtype==np.String or column.dtype==np.object:
        if column.dtype==np.object:
            ret[column_name] = convert_to_integer(column)
        else:
            ret[column_name] = column
    return ret
```

```python
data_agg = convert_strings_to_integer(data_agg)
data_agg.isnull().any()
```

```python
data_no_nan = data_agg.dropna()
data_no_nan.isnull().any()
```

```python
data_no_nan.head()
data_no_nan.shape
```

```python
from sklearn import ensemble
```

```python
# random-forest
n_sample=data_no_nan.shape[0] #1564
n_feature=data_no_nan.shape[1] #23

# Create the random forest object for the fit
forest = ensemble.RandomForestClassifier(n_estimators = 100)
# Fit the training data to the skin labels and create the decision trees
forest = forest.fit(data_no_nan.ix[0::, :22],data_no_nan.ix[0::, 23])

```

```python
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(n_feature):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```

```python
import matplotlib.pyplot as plt
# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(n_feature), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(n_feature), indices)
plt.xlim([-1, n_feature])
plt.show()
```

```python
# k-fold cross-validation
from sklearn.model_selection import KFold
kf = KFold(n_sample, n_folds=10, shuffle=False)
```
