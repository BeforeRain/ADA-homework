```python
uni_canton=pd.DataFrame(columns=['University','Canton'])
```

```python
def cantonInCH(result):
    for i in range(len(result)):
        if(result['geonames'][i]['countryName']=='Switzerland'):
            return True
    return False
```

```python
def cantonGet(uni):
    paraload={'q':uni,'username':'shiyuenie'}
    r = requests.get(url,params=paraload)
    result=r.json()
    return result
```

```python
import json
import pandas as pd

CH_GEO_JSON = 'ch-cantons-lg.topojson'

with open(CH_GEO_JSON, 'r') as f:
    d = json.load(f)

cantons = d['objects']['cantons']['geometries']
canton_id_name= []
for canton in cantons:
    canton_id_name.append((canton['id'],canton['properties']['name']))
    
canton_id_name
#canton_id_name = pd.DataFrame(canton_id_name, columns=['canton_ids', 'canton_names']) 
#canton_id_name.head()
```

```python
german_canton = ["Aargau", "Appenzell Ausserrhoden", "Appenzell Innerrhoden", "Basel-Stadt", 
                 "Basel-Landschaft", "Glarus", "Lucerne", "Nidwalden", "Obwalden", "Schaffhausen", 
                 "Schwyz", "Solothurn", "St. Gallen", "Thurgau", "Uri", "Zug", "Zurich"]
# https://en.wikipedia.org/wiki/Languages_of_Switzerland
# In 17 Swiss cantons, German is the only official language (Aargau, Appenzell Ausserrhoden, Appenzell Innerrhoden, Basel-Stadt, Basel-Landschaft, Glarus, Lucerne, Nidwalden, Obwalden, Schaffhausen, Schwyz, Solothurn, St. Gallen, Thurgau, Uri, Zug, Zurich).
```

```python
gr_co_fr_canton = ["Bern", "Fribourg", "Valais"] 
# Bern (German-speaking majority), Valais (French-speaking majority), and Fribourg (French-speaking majority)
tri_canton = ["Graubünden"] # more than half German
# In the cantons of Bern, Fribourg and Valais, French is co-official; in the trilingual canton of Graubünden, 
# more than half of the population speaks German, while the rest speak Romansh or Italian.
french_canton = ["Geneva", "Vaud", "Neuchâtel", "Jura"]
italian_canton = ["Ticino"]
```

```python
# german: gr  french: fr  italian: it   romansh: ro
canton_lg=[]
canton_nf=[]
for i in range(len(canton_id_name)):
    lg_flag=0
    c_name=canton_id_name[i][1]
    for gi in range (len(german_canton)):
        if (c_name==german_canton[gi]):
            canton_lg.append((canton_id_name[i][0],'gr'))
            cc[i]['code']=cc[i]['id']
            cc[i]['id']='gr'
            lg_flag+=1
    for fi in range (len(french_canton)):
        if (c_name==french_canton[fi]):
            canton_lg.append((canton_id_name[i][0],'fr'))
            cc[i]['code']=cc[i]['id']
            cc[i]['id']='fr'
            lg_flag+=1
    for gf in range (len(gr_co_fr_canton)):
        if (c_name==gr_co_fr_canton[gf]):
            canton_lg.append((canton_id_name[i][0],'gr/fr'))
            cc[i]['code']=cc[i]['id']
            cc[i]['id']='gr/fr'
            lg_flag+=1
    if(c_name=="Ticino"):
        canton_lg.append((canton_id_name[i][0],'it'))
        cc[i]['code']=cc[i]['id']
        cc[i]['id']='it'
        lg_flag+=1
    if(c_name=="Graubünden"):   
        canton_lg.append((canton_id_name[i][0],'gr/it/ro'))
        cc[i]['code']=cc[i]['id']
        cc[i]['id']='gr/it/ro'
        lg_flag+=1
    if (lg_flag==0):
        canton_nf.append((i,c_name,canton_id_name[i][0]))

canton_lg,canton_nf
```
